```python
import numpy as np
from itertools import product
import pandas as pd
import math
from scipy.special import factorial
```

```python
a =[]
answer = 0.71428571428
answer_list = []
ranges = [
    [0, 1],     # a2
    [0, 1, 2],  # a3
    [0, 1, 2, 3],  # a4
    [0, 1, 2, 3, 4],  # a5
    [0, 1, 2, 3, 4, 5],  # a6
    [0, 1, 2, 3, 4, 5, 6]  # a7
]

# Generate all combinations
combinations = list(product(*ranges))

# Filter out combinations with no repeating values
valid_combinations = [[*combo] for combo in combinations if len(set(combo)) == len(combo)]

# # Print the valid combinations
# for combo in valid_combinations:
#     print(combo)
print(f"Number f valid combinations :{len(valid_combinations)}")

for combo in valid_combinations:
    ans = (combo[0]/factorial(2)) + (combo[1]/factorial(3)) + (combo[2]/factorial(4)) + (combo[3]/factorial(5)) + (combo[4]/factorial(6)) + (combo[5]/factorial(7))
    answer_list.append(ans)

# print(len(answer_list))
df = pd.DataFrame({"Combo": valid_combinations, "answer": answer_list})

for i, row in df.iterrows():
    print(row)
    if math.isclose(row["answer"], answer):
        print(row)
```
